```python
courseDictList = []
for ind_class in set(relevant_courses):
#     print('https://fireroad.mit.edu/courses/lookup/' + ind_class)
    x = requests.get('https://fireroad.mit.edu/courses/lookup/' + ind_class)
    courseDictList.append(json.loads(x.text))
```

```python
toPandas = {}
for i in courseDictList:
    if(i['subject_id'] not in toPandas):
        toPandas[i['subject_id']] = {}
    try:
        toPandas[i['subject_id']]['hours'] = i['out_of_class_hours'] + i['in_class_hours']
    except:
        if i['total_units'] <= 0:
            toPandas[i['subject_id']]['hours'] = 12
        else:
            toPandas[i['subject_id']]['hours'] = i['total_units']

    try:
        toPandas[i['subject_id']]['rating'] = i['rating']
    except:
        toPandas[i['subject_id']]['rating'] = 5.5
        
    toPandas[i['subject_id']]['units'] = i['total_units']
    toPandas[i['subject_id']]['offered_fall'] = i['offered_fall']
    toPandas[i['subject_id']]['offered_spring']= i['offered_spring']
#     try:
#         toPandas[i['subject_id']]['prerequisites']= i['prerequisites']
#     except:
#         pass
    try:
        toPandas[i['subject_id']]['prerequisites']= i['prerequisites']
    except:
        pass
    
    try:
        toPandas[i['subject_id']]['is_hass']= i['hass_attribute'] != ''
    except:
        toPandas[i['subject_id']]['is_hass']= False
    
    try:
        toPandas[i['subject_id']]['schedule']= i['schedule']
    except:
        pass
```
